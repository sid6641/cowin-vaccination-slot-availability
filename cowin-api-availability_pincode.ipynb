```python
import requests
import datetime, time
import json
from notify_run import Notify 
```

```python
# 560001 560003 and 560017
# POST_CODE = "400708"
age = 26

# Print details flag
print_flag = 'Y'

numdays = 26
```

```python
base = datetime.datetime.today()
date_list = [base + datetime.timedelta(days=x) for x in range(numdays)]
date_str = [x.strftime("%d-%m-%Y") for x in date_list]
```

```python
def alertme(centername, date, pincode):
    print("sid : vaccine found !!!")
    notify = Notify() 
    notifyStr = f"ALERT: VACCINES ARE OPEN, BOOK NOW !!! pincode - {pincode}, {centername}, on date - {date}"
    notify.send(notifyStr) 
    
#     560001 -> 560040 -> all bangalore pincodes

all_pincodes = list(range(560001, 560041)) 
all_pincodes_str = [str(i) for i in all_pincodes]

while (True):
    for POST_CODE in all_pincodes_str:
        print(f"Checking for pincode - {POST_CODE}")
        for INP_DATE in date_str:
            URL = "https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/calendarByPin?pincode={}&date={}".format(POST_CODE, INP_DATE)
            response = requests.get(URL)
            if response.ok:
                resp_json = response.json()
                # print(json.dumps(resp_json, indent = 1))
                flag = False
                if resp_json["centers"]:
                    print("Available on: {}".format(INP_DATE))
                    if(print_flag=='y' or print_flag=='Y'):
                        for center in resp_json["centers"]:
                            for session in center["sessions"]:
                                if session["min_age_limit"] <= age:
                                    print("\t", center["name"])
                                    print("\t", center["block_name"])
                                    print("\t Price: ", center["fee_type"])
                                    print("\t Available Capacity: ", session["available_capacity"])
                                    if session["available_capacity"] > 0:
                                        alertme(center["name"], INP_DATE, POST_CODE)
                                    if(session["vaccine"] != ''):
                                        print("\t Vaccine: ", session["vaccine"])
                                    print("\n\n")

    #                                 print(center["sessions"])



#                 else:
#                     print("No available slots on {}".format(INP_DATE))
            
    print("sid : Sleeping for 60 sec")
    time.sleep(60)
```
